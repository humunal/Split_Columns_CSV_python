```python
#df = pd.DataFrame(columns = ["sku","alchool","alt_image","animal","area","australian_like","awards","blended_grapes","bottleml","box_size","cellar_practice","country_of_manufacture","cost","custom_block","custom_label","filtration","florealnew","fruit","international_awards","is_featured","kind","kind_byo_organic","main_grape","manufacturer","method","mineral","myob_btb","myob_btc_h","oak1","pairing","price","product_name","region","second_fermentation","sparkling_residual_sugar","spontaneous_fermentation1","skin_contact","sku_short","spice","stripe_sub_enabled","sw_featured","tastingnotes","vegetal","vineyard_management","Vntg","wild_ferment","wine_producer"], dtype=object)

with open("att-start-1.csv") as f:
    # use the first line as column names in the dataframe
    col_names = f.readline()
    df = pd.DataFrame(columns = col_names.split(","), dtype=object)
    print(col_names)

    for line in f:
        d = {}
        # lines have more than 2 columns, but the trailing values are empty
        # so the format is col1, large_col2,...
        col1, *col2 = line.split(",")
        d["sku"] = col1
        for item in col2:
            try:
                if item.strip(): #disregard the empty trailing columns
                    k,v = item.split("=")
                    # we split on comma, so have occasional
                    k = k.strip('"')
                    v = v.strip('"')
                    d[k] = v
            except ValueError as e:
                # there is a column value with missing key
                print("Could not assign to column:", d["sku"], item)
                
        for col in df.columns:   # add potentially missing columns as None
            if col not in d:
                d[col] = None
        df = df.append(d, ignore_index = True)

print(df)
```

```python
df.to_csv("data_parsed.csv") #save
```
